```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,15,13,18,16]
plt.plot(x,y)
plt.title("Basic Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()  # it enables the grid
plt.plot(x,linestyle='dotted',color='green')
plt.plot(y,linestyle='dotted',color='red')
plt.plot(x,y,linestyle='dotted',color='blue')
```

```python
# BAR GRAPH

import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y)
plt.show()
```

```python
# HORIZONTAL BAR GRAPH

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
# SCATTER GRAPHS

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99, 86 , 87 , 88, 111, 86, 103, 87,
94, 78, 77, 85, 86])
plt.scatter(x, y)
plt.show()
```

```python
# HISTOGRAM

import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
# PIE CHART

import numpy as np
import matplotlib.pyplot as plt

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
lab = ['A', 'B', 'C', 'D', 'E', 'F', 'G','H']
c = ['green', 'yellow', 'red', 'blue', 'magenta', 'orange', 'lime', 'indigo']
plt.pie(y, labels=lab, colors=c, autopct="%0.2f%%")
plt.title('Pie Chart Example')
plt.show()
```
